# Comparative Analysis of DQN Variants in LunarLander-v2

This notebook presents a comparative analysis of different Deep Q-Network (DQN) variants, including Standard DQN, Double DQN, Noisy DQN, and a Combined DQN (Double & Noisy) approach. 

## Import Libarries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Loading Data
Loading the average score data from previous runs of Standard, Double, Noisy, and Combined DQN models.

```python
# Load the data
average_scores_standard = np.load('average_scores/average_scores_standard.npy')
average_scores_double = np.load('average_scores/average_scores_double.npy')
average_scores_noisy = np.load('average_scores/average_scores_noisy.npy')
average_scores_combined = np.load('average_scores/average_scores_combined.npy')
```

## Standard DQN Visualisation

```python
# Creating the plot for Standard DQN

# Generate a range of episode numbers for the x-axis
# Start from episode 10 and increment by 10 until the number of episodes recorded
episode_numbers_standard = list(range(1, len(average_scores_standard) + 1))

plt.figure(figsize=(12, 6))

# Plot average rewards per episode
plt.plot(episode_numbers_standard, average_scores_standard, label='Average Reward Standard DQN', color='blue')

# Add a horizontal line representing the solved threshold
plt.axhline(y=195, color='green', linestyle='--', label='Solved Threshold (195)')

# Set x-axis labels to show every 100 episodes
plt.xticks(range(0, len(average_scores_standard) + 1, 100))

# Set labels for x and y axes
plt.xlabel('Episodes')
plt.ylabel('Average Reward')

# Display the plot
plt.title('Average Reward per Episode in LunarLander-v2 (Standard DQN)')
plt.legend()
plt.show()
```

```python
# Creating the plot for Double DQN

# Generate a range of episode numbers for the x-axis
# Start from episode 10 and increment by 10 until the number of episodes recorded
episode_numbers_double = list(range(1, len(average_scores_double) + 1))

plt.figure(figsize=(12, 6))

# Plot average rewards per episode
plt.plot(episode_numbers_double, average_scores_double, label='Average Reward Double DQN', color='red')

# Add a horizontal line representing the solved threshold
plt.axhline(y=195, color='green', linestyle='--', label='Solved Threshold (195)')

# Set x-axis labels to show every 100 episodes
plt.xticks(range(0, len(average_scores_double) + 1, 100))

# Set labels for x and y axes
plt.xlabel('Episodes')
plt.ylabel('Average Reward')

# Display the plot
plt.title('Average Reward per Episode in LunarLander-v2 (Standard DQN)')
plt.legend()
plt.show()
```

## Comparative Analysis

Plotting all DQN variants into a single plot to evaluate their relative performance and efficiency in solving the LunarLander-v2 environment.

```python
# Generate episode numbers for each model
episode_numbers_standard = list(range(1, len(average_scores_standard) + 1))
episode_numbers_double = list(range(1, len(average_scores_double) + 1))
episode_numbers_noisy = list(range(1, len(average_scores_noisy) + 1))
episode_numbers_combined = list(range(1, len(average_scores_combined) + 1))

# Create the plot
plt.figure(figsize=(12, 6))

plt.plot(episode_numbers_standard, average_scores_standard, label='Standard DQN', color='blue')
plt.plot(episode_numbers_double, average_scores_double, label='Double DQN', color='red')
plt.plot(episode_numbers_noisy, average_scores_noisy, label='Noisy DQN', color='orange')
plt.plot(episode_numbers_combined, average_scores_combined, label='Combined DQN', color='pink')  

# Add a horizontal line for the solved threshold
plt.axhline(y=195, color='gray', linestyle='--', label='Solved Threshold')

plt.xlabel('Episodes')
plt.ylabel('Average Reward')
plt.title('Comparison of DQN Models in LunarLander-v2')
plt.legend()
plt.show()
```
