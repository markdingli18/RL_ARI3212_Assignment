### Import Libarries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
average_scores_stochastic = np.load('average_scores_experiment_2/average_scores_stochastic.npy')
average_scores_deterministic = np.load('average_scores_experiment_2/average_scores_deterministic.npy')
```

### PPO-GAE (Stochastic)

```python
# Creating the plot for PPO-GAE (Stochastic)

# Generate a range of episode numbers for the x-axis
# Start from episode 10 and increment by 10 until the number of episodes recorded
episode_numbers_standard = list(range(1, len(average_scores_stochastic) + 1))

plt.figure(figsize=(12, 6))

# Plot average rewards per episode
plt.plot(episode_numbers_standard, average_scores_stochastic, label='Average Reward PPO-GAE (Stochastic)', color='blue')

# Add a horizontal line representing the solved threshold
plt.axhline(y=195, color='green', linestyle='--', label='Solved Threshold (195)')

# Set x-axis labels to show every 100 episodes
plt.xticks(range(0, len(average_scores_stochastic) + 1, 100))

# Set labels for x and y axes
plt.xlabel('Episodes')
plt.ylabel('Average Reward')

# Display the plot
plt.title('Average Reward per Episode in LunarLanderContinuous-v2 (PPO-GAE)')
plt.legend()
plt.show()
```

### DDPG (Deterministic)

```python
# Creating the plot for DDPG (Deterministic)

# Generate a range of episode numbers for the x-axis
# Start from episode 10 and increment by 10 until the number of episodes recorded
episode_numbers_standard = list(range(1, len(average_scores_deterministic) + 1))

plt.figure(figsize=(12, 6))

# Plot average rewards per episode
plt.plot(episode_numbers_standard, average_scores_deterministic, label='Average Reward DDPG (Deterministic)', color='red')

# Add a horizontal line representing the solved threshold
plt.axhline(y=195, color='green', linestyle='--', label='Solved Threshold (195)')

# Set x-axis labels to show every 100 episodes
plt.xticks(range(0, len(average_scores_deterministic) + 1, 100))

# Set labels for x and y axes
plt.xlabel('Episodes')
plt.ylabel('Average Reward')

# Display the plot
plt.title('Average Reward per Episode in LunarLanderContinuous-v2 (DDPG)')
plt.legend()
plt.show()
```

### Comparisons

```python
# Creating the combined plot for PPO-GAE (Stochastic) and DDPG (Deterministic)

plt.figure(figsize=(12, 6))

# Plot for PPO-GAE (Stochastic)
episode_numbers_stochastic = list(range(1, len(average_scores_stochastic) + 1))
plt.plot(episode_numbers_stochastic, average_scores_stochastic, label='Average Reward PPO-GAE (Stochastic)', color='blue')

# Plot for DDPG (Deterministic)
episode_numbers_deterministic = list(range(1, len(average_scores_deterministic) + 1))
plt.plot(episode_numbers_deterministic, average_scores_deterministic, label='Average Reward DDPG (Deterministic)', color='red')

# Add a horizontal line representing the solved threshold
plt.axhline(y=195, color='green', linestyle='--', label='Solved Threshold (195)')

# Set x-axis labels to show every 100 episodes (adjust according to your dataset size)
plt.xticks(range(0, max(len(average_scores_stochastic), len(average_scores_deterministic)) + 1, 100))

# Set labels for x and y axes
plt.xlabel('Episodes')
plt.ylabel('Average Reward')

# Add title and legend
plt.title('Comparison of Average Reward per Episode in LunarLanderContinuous-v2')
plt.legend()

# Display the plot
plt.show()
```
